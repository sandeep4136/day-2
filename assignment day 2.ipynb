```python
import math
(x,y)=(0,0)

for i in range(int(input())):
    a,b=input().split()
    a,b=str(a),int(b)
    if a=="DOWN":x,y=x,y-b
    elif a=="LEFT":x,y=x-b,y
    elif a=="RIGHT":x,y=x+b,y
    else:x,y=x,y+b
print(round(math.sqrt((x**2)+(y**2))))
```
